```python
class BinarySearchTree:
    # Código anterior da árvore (inserção, etc.) fica aqui...

    def min_value(self):
        if self.root is None:
            return None

        current = self.root
        while current.left is not None:
            current = current.left

        return current.value
```

```python
class BinarySearchTree:
    # Código anterior da árvore (inserção, etc.) fica aqui...

    def max_value(self):
        if self.root is None:
            return None

        current = self.root
        while current.right is not None:
            current = current.right

        return current.value
```

```python
class BinarySearchTree:
    # Código anterior da árvore aqui (Node, insert, etc.)

    def delete(self, value):
        self.root = self._delete_recursive(self.root, value)

    def _delete_recursive(self, node, value):
        if node is None:
            return None

        # Busca o nó a ser removido
        if value < node.value:
            node.left = self._delete_recursive(node.left, value)
        elif value > node.value:
            node.right = self._delete_recursive(node.right, value)
        else:
            # Caso 1: Nó folha
            if node.left is None and node.right is None:
                return None

            # Caso 2: Um filho
            elif node.left is None:
                return node.right
            elif node.right is None:
                return node.left

            # Caso 3: Dois filhos
            else:
                min_larger_node = self._min_value_node(node.right)
                node.value = min_larger_node.value
                node.right = self._delete_recursive(node.right, min_larger_node.value)

        return node

    def _min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current
```

```python
class BinarySearchTree:
    # Código anterior da classe (Node, insert, delete etc.)

    def height(self):
        return self._height_recursive(self.root)

    def _height_recursive(self, node):
        if node is None:
            return -1  # Árvore vazia tem altura -1

        left_height = self._height_recursive(node.left)
        right_height = self._height_recursive(node.right)

        return 1 + max(left_height, right_height)
```

```python
class BinarySearchTree:
    # Código anterior da classe (Node, insert, delete, etc.)

    def pre_order_traversal(self):
        elements = []
        self._pre_order_recursive(self.root, elements)
        return elements

    def _pre_order_recursive(self, node, elements):
        if node:
            elements.append(node.value)                  # Visita o nó atual
            self._pre_order_recursive(node.left, elements)   # Percorre a esquerda
            self._pre_order_recursive(node.right, elements)  # Percorre a direita
```
